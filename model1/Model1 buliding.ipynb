```python
def set_processor(name):
    if name in ['Intel Core i7', 'Intel Core i5', 'Intel Core i3']:
        return name  # Return actual name instead of string 'name'
    elif name.split()[0] == 'AMD':
        return 'AMD'
    else:
        return 'Other'  # Capitalized 'Other' for consistency

data['cpu_name'] = data['cpu_name'].apply(set_processor)

        
```

```python
# Function to categorize OS
def set_os(inpt):
    if inpt == 'Windows 10': 
        return 'Windows'
    elif inpt == 'macOS':
        return 'Mac'
    elif inpt == 'Mac OS X':
        return 'Mac'
    elif inpt == 'Linux':
        return 'Linux'
    elif inpt == 'Windows 10 S':
         return 'Windows'
    
    else:
        return 'Other'

# Apply the function to the OpSys column
data.loc[:, 'OpSys'] = data['OpSys'].apply(set_os)
```

```python
data['OpSys'] .value_counts()
```

```python
data=data.drop(columns= ['laptop_ID','Inches','Product','ScreenResolution','Cpu','Gpu'])
```

```python
data.head()
```

```python
data= pd.get_dummies(data)
```

```python
data.head()
```

```python
data.shape
```

```python
X= data.drop('Price_euros',axis=1)
y= data['Price_euros']
```

```python
!pip install scikit-learn
!pip install --upgrade pip
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
X_train.shape,X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train,y_train)
    acc=model.score(X_test,y_test)
    print(str(model)+ ' -->' +str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr =LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
lasso=Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {
    'n_estimators': [10, 50, 100],
    'criterion': ['squared_error', 'absolute_error', 'poisson']
}

# Define the model
rf = RandomForestRegressor()

# Perform Grid Search
grid_obj = GridSearchCV(estimator=rf, param_grid=parameters, cv=5)  # Add cv for cross-validation
grid_fit = grid_obj.fit(X_train, y_train)  # Make sure X_train and y_train are defined

# Get the best model
best_model = grid_fit.best_estimator_
best_model
```

```python
best_model
```

```python
best_model.score(X_test,y_test)
```
